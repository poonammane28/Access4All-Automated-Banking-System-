```python
import speech_recognition as sr
import pyttsx3
from googletrans import Translator
import pytesseract
import cv2

# Initialize the text-to-speech engine
tts_engine = pyttsx3.init()
translator = Translator()

# Set Tesseract OCR path (if needed, e.g., on Windows)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Function to speak the response
def speak(text):
    tts_engine.say(text)
    tts_engine.runAndWait()

# Function to process the command
def process_command(command):
    command = command.lower()

    if "balance" in command:
        return "check_balance"
    elif "transfer" in command:
        return "transfer_money"
    elif "statement" in command:
        return "generate_statement"
    elif "deposit" in command:
        return "deposit_money"
    elif "withdraw" in command:
        return "withdraw_money"
    elif "update" in command and "information" in command:
        return "update_personal_info"
    elif "block" in command and "card" in command:
        return "block_card"
    elif "activate" in command and "card" in command:
        return "activate_card"
    elif "loan" in command:
        return "loan_services"
    elif "open" in command and "account" in command:
        return "open_account"
    elif "close" in command and "account" in command:
        return "close_account"
    elif "cheque" in command:
        return "cheque_services"
    elif "help" in command or "support" in command:
        return "customer_support"
    elif "limit" in command and "transaction" in command:
        return "set_transaction_limit"
    elif "password" in command or "pin" in command:
        return "change_password_or_pin"
    else:
        return "unknown"

# Function to scan and read the passbook
import os

# Function to scan and read the passbook
def scan_passbook():
    image_path = input("Enter the path to the passbook image: ")

    # Ensure backslashes are treated correctly
    image_path = image_path.strip().replace("\\", "/")  

    # Check if the file exists
    if not os.path.exists(image_path):
        print("Error: File does not exist. Please provide a valid path.")
        return

    # Load the image
    img = cv2.imread(image_path)

    if img is None:
        print("Error: Unable to load the image. Ensure the file is an image and the path is correct.")
        return

    # Convert to grayscale for better OCR results
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Perform OCR on the image
    try:
        text = pytesseract.image_to_string(gray)
        print("\n--- Extracted Text ---")
        print(text)
        print("-----------------------")
        speak("Passbook scanned successfully. Details are displayed on the screen.")
    except Exception as e:
        print(f"Error during OCR: {e}")
        speak("An error occurred while processing the passbook.")

# Function to perform the activity
def perform_activity(action):
    if action != "unknown":
        speak("Would you like to scan your passbook? Please answer yes or no.")
        scan_choice = input("Would you like to scan your passbook? (yes/no): ").lower()

        if scan_choice == "yes":
            scan_passbook()
            speak(f"The action {action.replace('_', ' ')} will be performed.")
        else:
            speak("Passbook scanning skipped.")
    else:
        speak("I'm sorry, I didn't understand your request.")

# Function to recognize speech and translate it to English
def recognize_speech(language_code):
    recognizer = sr.Recognizer()
    try:
        with sr.Microphone() as source:
            speak(f"Listening in {language_code}. Please speak now.")
            print("Listening...")
            audio = recognizer.listen(source)
            speech_text = recognizer.recognize_google(audio, language=language_code)
            print(f"You said: {speech_text}")

            # Translate to English
            translated_text = translator.translate(speech_text, src=language_code, dest="en").text
            print(f"Translated to English: {translated_text}")

            # Process the translated command
            action = process_command(translated_text)
            perform_activity(action)

    except sr.UnknownValueError:
        speak("Sorry, I couldn't understand what you said. Please try again.")
    except sr.RequestError as e:
        speak(f"Could not process your request; {e}")

# Main function to select language and process command
def main():
    language_map = {
        "1": "hi",  # Hindi
        "2": "ta",  # Tamil
        "3": "te",  # Telugu
        "4": "kn",  # Kannada
        "5": "mr",  # Marathi
    }

    print("Select a language:")
    for key, value in language_map.items():
        print(f"{key}: {value}")

    choice = input("Enter the number corresponding to your language: ")
    language_code = language_map.get(choice, "en")  # Default to English
    recognize_speech(language_code)

# Run the program
if __name__ == "__main__":
    main()
```

```python
import os
import re
import speech_recognition as sr
import pyttsx3
from googletrans import Translator
import pytesseract
import cv2
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Initialize the text-to-speech engine
tts_engine = pyttsx3.init()
translator = Translator()

# Set Tesseract OCR path (if needed, e.g., on Windows)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Function to speak the response
def speak(text):
    tts_engine.say(text)
    tts_engine.runAndWait()

# Function to process the command
def process_command(command):
    command = command.lower()

    if "balance" in command:
        return "check_balance"
    elif "transfer" in command:
        return "transfer_money"
    elif "statement" in command:
        return "generate_statement"
    elif "deposit" in command:
        return "deposit_money"
    elif "withdraw" in command:
        return "withdraw_money"
    elif "update" in command and "information" in command:
        return "update_personal_info"
    elif "block" in command and "card" in command:
        return "block_card"
    elif "activate" in command and "card" in command:
        return "activate_card"
    elif "loan" in command:
        return "loan_services"
    elif "open" in command and "account" in command:
        return "open_account"
    elif "close" in command and "account" in command:
        return "close_account"
    elif "cheque" in command:
        return "cheque_services"
    elif "help" in command or "support" in command:
        return "customer_support"
    elif "limit" in command and "transaction" in command:
        return "set_transaction_limit"
    elif "password" in command or "pin" in command:
        return "change_password_or_pin"
    else:
        return "unknown"

# Function to scan and read the passbook
def scan_passbook():
    image_path = input("Enter the path to the passbook image: ")

    # Ensure backslashes are treated correctly
    image_path = image_path.strip().replace("\\", "/")  

    # Check if the file exists
    if not os.path.exists(image_path):
        print("Error: File does not exist. Please provide a valid path.")
        return None

    # Load the image
    img = cv2.imread(image_path)

    if img is None:
        print("Error: Unable to load the image. Ensure the file is an image and the path is correct.")
        return None

    # Convert to grayscale for better OCR results
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Perform OCR on the image
    try:
        text = pytesseract.image_to_string(gray)
        print("\n--- Extracted Text ---")
        print(text)
        print("-----------------------")
        speak("Passbook scanned successfully. Details are displayed on the screen.")
        return text
    except Exception as e:
        print(f"Error during OCR: {e}")
        speak("An error occurred while processing the passbook.")
        return None

# Function to extract fields from passbook text
def extract_fields(text):
    fields = {
        'Name': re.search(r"Name[:\- ]*(.*)", text, re.IGNORECASE),
        'Account Number': re.search(r"Account Number[:\- ]*(\d+)", text, re.IGNORECASE),
        'Branch': re.search(r"Branch[:\- ]*(.*)", text, re.IGNORECASE)
    }
    for key, match in fields.items():
        fields[key] = match.group(1).strip() if match else "Not Found"
    return fields

# Function to generate and autofill the form
def generate_and_autofill_form(action, fields):
    form_file = "Autofilled_Form.pdf"
    c = canvas.Canvas(form_file, pagesize=letter)
    c.drawString(100, 750, f"Form Type: {action.replace('_', ' ').title()}")
    c.drawString(100, 720, f"Name: {fields.get('Name', 'Not Found')}")
    c.drawString(100, 700, f"Account Number: {fields.get('Account Number', 'Not Found')}")
    c.drawString(100, 680, f"Branch: {fields.get('Branch', 'Not Found')}")
    c.drawString(100, 660, "Additional Details: [Fill as required]")
    c.save()
    speak("Form generated successfully.")
    print(f"Form saved as {form_file}. Please check your directory.")

# Function to perform the activity
def perform_activity(action, passbook_text):
    if action == "unknown":
        speak("I'm sorry, I didn't understand your request.")
        return

    speak(f"The action {action.replace('_', ' ')} will be performed.")
    if passbook_text:
        fields = extract_fields(passbook_text)
        generate_and_autofill_form(action, fields)
    else:
        speak("Passbook details are unavailable. Unable to autofill the form.")

# Function to recognize speech and translate it to English
def recognize_speech(language_code):
    recognizer = sr.Recognizer()
    try:
        with sr.Microphone() as source:
            speak(f"Listening in {language_code}. Please speak now.")
            print("Listening...")
            audio = recognizer.listen(source)
            speech_text = recognizer.recognize_google(audio, language=language_code)
            print(f"You said: {speech_text}")

            # Translate to English
            translated_text = translator.translate(speech_text, src=language_code, dest="en").text
            print(f"Translated to English: {translated_text}")

            # Process the translated command
            action = process_command(translated_text)
            speak("Would you like to scan your passbook? Please answer yes or no.")
            scan_choice = input("Would you like to scan your passbook? (yes/no): ").lower()

            if scan_choice == "yes":
                passbook_text = scan_passbook()
            else:
                passbook_text = None

            perform_activity(action, passbook_text)

    except sr.UnknownValueError:
        speak("Sorry, I couldn't understand what you said. Please try again.")
    except sr.RequestError as e:
        speak(f"Could not process your request; {e}")

# Main function to select language and process command
def main():
    language_map = {
        "1": "hi",  # Hindi
        "2": "ta",  # Tamil
        "3": "te",  # Telugu
        "4": "kn",  # Kannada
        "5": "mr",  # Marathi
    }

    print("Select a language:")
    for key, value in language_map.items():
        print(f"{key}: {value}")

    choice = input("Enter the number corresponding to your language: ")
    language_code = la1
    nguage_map.get(choice, "en")  # Default to English
    recognize_speech(language_code)

# Run the program
if __name__ == "__main__":
    main()
```
